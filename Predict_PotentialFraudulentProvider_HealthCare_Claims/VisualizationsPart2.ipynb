```python
# First, we'll import pandas, a data processing and CSV file I/O library
import pandas as pd

# We'll also import seaborn, a Python graphing library
import warnings 
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)

# Next, we'll load the FinalMergeData dataset
data = pd.read_csv("FinalMergeData.csv") 

# Let's check the data
data.head()

```

```python
#Load the original providers with target data
providerfraud=pd.read_csv("Train-1542865627584.csv") 
```

```python
target_count1 = providerfraud['PotentialFraud'].value_counts()
print('Class No:', target_count1[0])
print('Class Yes:', target_count1[1])
print('Proportion:', round(target_count1[1] / target_count1[0], 2), ': 1')

target_count1.plot(kind='bar', title='Count (PotentialFraud)');
```

Observations:There is class imbalance,there are 506 providers which are fraudulent out of 5410 providers

```python
target_count = data['PotentialFraud'].value_counts()
print('Class No:', target_count[0])
print('Class Yes:', target_count[1])
print('Proportion:', round(target_count[1] / target_count[0], 2), ': 1')

target_count.plot(kind='bar', title='Count (PotentialFraud)');
```

Observations:There is class imbalance in the data with proportions 0.62:1

```python
# We'll use seaborn's FacetGrid to color the scatterplot by PotentialFraud
sns.FacetGrid(data, hue="PotentialFraud", size=5) \
   .map(plt.scatter, "DaysAdmitted", "ClaimDuration") \
   .add_legend()
```

Plot Explaination:
We have plotted the newly generated features DaysAdmitted vs ClaimDuration wrt target classes(PotentialFraud)

Observations:
We see the Number of fraudulant claims are more when ClaimDurations are greater than 0 and No.of days admitted=0(That is for an inpatient ClaimDuration being of a higher value can be an indicator for fraud)


```python
 data["ChronicCond_KidneyDisease"]=data["ChronicCond_KidneyDisease"].astype('category')
g = sns.catplot(x="ChronicCond_KidneyDisease", hue="PotentialFraud", col="RenalDiseaseIndicator",
                data=data, kind="count",
                 height=4, aspect=.7);
g
```

Plot Explaination:
We see the two features RenalDiseaseIndicator and ChronicCond_KidneyDisease will be dependent on each other
ie,if ChronicCond_KidneyDisease is 1 then RenalDiseaseIndicator should be Y and vice versa

Observation:
There are more chances for a claim to be fraud if the above condition is not satisfied.

```python
import seaborn as sns
corr = data.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

Observations:
Strong Correlations between (InscClaimAmtReimbursed,DeductibleAmtPaid) ,
(DaysAdmitted,InscClaimAmtReimbursed),(DaysAdmitted,DeductibleAmtPaid),(IPAnnualDeductibleAmt ,IPAnnualReimbursementAmt)


```python

# Bar Plot
fig = plt.figure(figsize = (6, 4))
title = fig.suptitle("Potential Fraud Frequency", fontsize=14)
fig.subplots_adjust(top=0.85, wspace=0.3)

ax = fig.add_subplot(1,1, 1)
ax.set_xlabel("PotentialFraud")
ax.set_ylabel("Frequency") 
w_q = data['PotentialFraud'].value_counts()
w_q = (list(w_q.index), list(w_q.values))
ax.tick_params(axis='both', which='major', labelsize=8.5)
bar = ax.bar(w_q[0], w_q[1], color='steelblue', 
        edgecolor='black', linewidth=1)
```

```python
# Correlation Matrix Heatmap
f, ax = plt.subplots(figsize=(10, 6))
corr = data.corr()
hm = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=.05)
f.subplots_adjust(top=0.93)
t= f.suptitle('Merged Data Attributes Correlation Heatmap', fontsize=14)
```
